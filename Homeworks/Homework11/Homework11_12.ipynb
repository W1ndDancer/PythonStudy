Ниже я написал немного кода, для обощения информации по всем континентам и по миру в целом.

По этим данным мы видим что в мире на данный момент проживает примерно 7,9 млрд человек и что в день рождается 236 новых людей.

Конкретные данные по континентам я проанализирую дальше на графиках, чтоб было нагляднее, но при желании вы можете их изучить и в таком виде.



```python
dataAfr = df[df['Continent'] == 'Africa']
dataAsi = df[df['Continent'] == 'Asia']
dataEu = df[df['Continent'] == 'Europe']
dataNA = df[df['Continent'] == 'North America']
dataSA = df[df['Continent'] == 'South America']
dataOc = df[df['Continent'] == 'Oceania']

totworld1 = df[['2022 Population','Area (km²)','Growth Rate']].sum()
afr = dataAfr[['2022 Population','Area (km²)','Growth Rate', 'World Population Percentage']].sum()
asi = dataAsi[['2022 Population','Area (km²)','Growth Rate', 'World Population Percentage']].sum()
eu = dataEu[['2022 Population','Area (km²)','Growth Rate', 'World Population Percentage']].sum()
na = dataNA[['2022 Population','Area (km²)','Growth Rate', 'World Population Percentage']].sum()
sa = dataSA[['2022 Population','Area (km²)','Growth Rate', 'World Population Percentage']].sum()
oc = dataOc[['2022 Population','Area (km²)','Growth Rate', 'World Population Percentage']].sum()


print(f'World info:\n{totworld1}')
# Плотность населения = (количество людей на планете)/(площадь проживания) З.Ы. Сократил числа для удобства) 
print()
print(f'Africa:\n{afr}')
print()
print(f'Asia:\n{asi}')
print()
print(f'Europe:\n{eu}')
print()
print(f'North America:\n{na}')
print()
print(f'SouthAmerica:\n{sa}')
print()
print(f'Oceania:\n{oc}')

```

##Территория
На гистограмме ниже сравниваются общая территория континентов.

Из нее мы видим что самый больщой континет это Азия, на втором месте идет Африка, потом Северная Америка и остальные.

```python
df.loc[df['Continent'] == 'Africa', 'Cont_group'] = 'Africa'
df.loc[df['Continent'] == 'Asia', 'Cont_group'] = 'Asia'
df.loc[df['Continent'] == 'Europe', 'Cont_group'] = 'Europe'
df.loc[df['Continent'] == 'North America', 'Cont_group'] = 'NA'
df.loc[df['Continent'] == 'South America', 'Cont_group'] = 'SA'
df.loc[df['Continent'] == 'Oceania', 'Cont_group'] = 'Oceania'
df.groupby('Cont_group')['Area (km²)'].sum().plot(kind='bar')
```
